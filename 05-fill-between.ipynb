```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

### Read the data

```python
data = pd.read_csv('Data/new_data.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
# Average salary between all different developers.
int(dev_salaries.mean())
```

### Plot the data

```python
plt.plot(
    ages,
    dev_salaries,
    linestyle='--',
    label='All Devs',
    color='k',
)

plt.plot(
    ages,
    py_salaries,
    linestyle='-',
    label='Python',
    color='r'
)

overall_median = 69_257 #$

plt.fill_between(
    ages,
    py_salaries,
    overall_median,
    where=(py_salaries > overall_median),
    interpolate=True,
    color='green',
    alpha=0.50,
    label='Above Avg',
)

plt.fill_between(
    ages,
    py_salaries,
    overall_median,
    where=(py_salaries <= overall_median),
    interpolate=True,
    color='red',
    alpha=0.50,
    label='Below Avg',
)

plt.legend(loc='upper left')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
