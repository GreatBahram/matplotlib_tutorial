```python
# Built-in modules
from pathlib import Path


import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
from IPython.core.display import display, HTML

# Enable matplotlib inline diagraming.
%matplotlib inline

# Increase the width of Ipython notebook to 80%
display(HTML("<style>.container { width:80% !important; }</style>"))

DATA_DIR = Path('../Data')

plt.style.use('seaborn')
```

```python
data = pd.read_csv(DATA_DIR / 'subplots.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.dtypes
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
# Subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

# Stateful
ax1.plot(ages, py_salaries, label='Python')
ax1.plot(ages, js_salaries, label='JavaScript')
ax2.plot(
    ages, 
    dev_salaries,
    label='All Devs',
    linestyle='--',
    color='#444444'
)

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.savefig('together.png')
plt.show()
```

## Two different figures

```python
# Subplots
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

# Stateful
ax1.plot(ages, py_salaries, label='Python')
ax1.plot(ages, js_salaries, label='JavaScript')
ax2.plot(
    ages, 
    dev_salaries,
    label='All Devs',
    linestyle='--',
    color='#444444'
)

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
fig1.savefig('fig1.png')
fig2.savefig('fig2.png')
plt.show()
```
